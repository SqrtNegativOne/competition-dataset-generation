```python
import numpy as np
import pandas as pd
from sklearn.datasets import *
```

```python
random_state = 699
```

```python
n_samples = 70000
noise = 0.3
factor = 0.7

circles_X, circles_y = make_circles(n_samples = n_samples, noise = noise, factor = factor, random_state = random_state)
outer, inner = 0, 1
labels = np.full(n_samples, -1.0)
labels[0] = outer
labels[-1] = inner
```

```python
circles = pd.DataFrame(circles_X)
circles['y'] = circles_y
circles.columns = ["c_1", "c_2", "c_y"]
circles.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))
plt.scatter(
    circles_X[circles_y == outer, 0],
    circles_X[circles_y == outer, 1],
    color="navy",
    marker="s",
    lw=0,
    label="outer labeled",
    s=10,
)
plt.scatter(
    circles_X[circles_y == inner, 0],
    circles_X[circles_y == inner, 1],
    color="c",
    marker="s",
    lw=0,
    label="inner labeled",
    s=10,
)
plt.legend(scatterpoints=1, shadow=False, loc="center")
_ = plt.title("Raw data (2 classes=outer and inner)")
```

```python
n_samples = 70000
noise = 0.01
factor = 0.7

moons_X, moons_y = make_moons(n_samples = n_samples, noise = noise, random_state = random_state)
outer, inner = 0, 1
```

```python
moons = pd.DataFrame(moons_X)
moons['y'] = moons_y
moons.columns = ["m_1", "m_2", "m_y"]
moons.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))
plt.scatter(
    moons_X[moons_y == outer, 0],
    moons_X[moons_y == outer, 1],
    color="navy",
    marker="s",
    lw=0,
    label="outer labeled",
    s=10,
)
plt.scatter(
    moons_X[moons_y == inner, 0],
    moons_X[moons_y == inner, 1],
    color="c",
    marker="s",
    lw=0,
    label="inner labeled",
    s=10,
)
plt.legend(scatterpoints=1, shadow=False, loc="center")
_ = plt.title("Raw data (2 classes=outer and inner)")
```

```python
friedman1_X, friedman1_y = make_friedman1(n_samples = n_samples, noise = noise, random_state = random_state)
friedman2_X, friedman2_y = make_friedman2(n_samples = n_samples, noise = noise, random_state = random_state)
friedman3_X, friedman3_y = make_friedman3(n_samples = n_samples, noise = noise, random_state = random_state)
```

```python
F1 = pd.DataFrame(friedman1_X)
F1['y'] = friedman1_y
F1.columns = ["f1_"+str(i) for i in F1.columns]

F2 = pd.DataFrame(friedman2_X)
F2['y'] = friedman2_y
F2.columns = ["f2_"+str(i) for i in F2.columns]

F3 = pd.DataFrame(friedman3_X)
F3['y'] = friedman3_y
F3.columns = ["f3_"+str(i) for i in F3.columns]
display(F1.head())
display(F2.head())
display(F3.head())
```

```python
hastie_X, hastie_y = make_hastie_10_2(n_samples = n_samples, random_state = random_state)
```

```python
hastie = pd.DataFrame(hastie_X)
hastie['y'] = hastie_y
hastie.columns = ["h10_2_"+str(i) for i in hastie.columns]
hastie.head()
```

```python
mc_X, mc_y = make_classification(n_samples = n_samples, n_features = 10, n_informative = 7, n_redundant=1, n_classes = 4, random_state = random_state)
```

```python
mc = pd.DataFrame(mc_X)
mc['y'] = mc_y
mc.columns = ["mc_"+str(i) for i in mc.columns]
mc.head()
```

```python
main_con = pd.concat([circles, moons, F1, F2, F3, hastie, mc], axis = 1)
main_con.head()
```

```python
y_cols = main_con.columns[[i.endswith('y') for i in main_con.columns]]
y_cold = main_con[y_cols]
y_cold
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
y_cold = pd.DataFrame(norm.fit_transform(y_cold))
```

```python
from sklearn.decomposition import PCA, LatentDirichletAllocation

pca = PCA(n_components=5, random_state = random_state)
main_y = pca.fit_transform(y_cold)[:,4]
```

```python
pca.explained_variance_ratio_
```

```python
main_con['target'] = main_y
main_con = main_con.drop(columns = y_cols)
main_con
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
tr, te = train_test_split(main_con, test_size = 0.3)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(tr.iloc[:,:-1], tr.iloc[:,-1])
```

```python
xgb.predict(te.iloc[:,:-1])
```

```python
r2_score(te.iloc[:,-1], xgb.predict(te.iloc[:,:-1]))
```
